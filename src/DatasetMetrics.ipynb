```python
import json

from bokeh.plotting import figure
from bokeh.io import output_notebook, show
```

# Training set

```python
with open("../data/training_set.json", "r") as training_set_file:
    training_set = json.load(training_set_file)
```

```python
utterance_per_class_train = {}

for data in training_set:
    utterance_per_class_train[data.get("intent")] = (utterance_per_class_train[data.get("intent")] + 1 if data.get("intent") in utterance_per_class_train else 1)
```

```python
print(f'Nombre de phrases dans le training set : {len(training_set)}')
```

```python
output_notebook()

p = figure(x_range=list(utterance_per_class_train.keys()), title="Utterance per class in the training set")
p.vbar(x=list(utterance_per_class_train.keys()), top=list(utterance_per_class_train.values()), fill_color="#ff8c00")
show(p)
```
