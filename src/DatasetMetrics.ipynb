```python
import json

from bokeh.plotting import figure
from bokeh.io import output_notebook, show
```

```python
with open("../data/training_set.json", "r") as training_set_file:
    training_set = json.load(training_set_file)
```

```python
utterance_per_class = {}

for data in training_set:
    utterance_per_class[data.get("intent")] = (utterance_per_class[data.get("intent")] + 1 if data.get("intent") in utterance_per_class else 1)
```

```python
output_notebook()

p = figure(x_range=list(utterance_per_class.keys()), title="Utterance per class")
p.vbar(x=list(utterance_per_class.keys()), top=list(utterance_per_class.values()), fill_color="#b3de69")
show(p)
```
